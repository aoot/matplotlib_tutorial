### Historgrams

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left', 'Right']
colors=['red', 'blue']

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f%%')
plt.title('Foot Preference of FIFA Players')
```

### Weights of the FIFA players

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]   ### Interesting Syntax

plt.style.use('ggplot') # Change color scheme

light = fifa.loc[fifa['Weight']<125].count()[0]
light_medium = fifa.loc[(fifa['Weight']>=125) & (fifa['Weight']<150)].count()[0]
medium = fifa.loc[(fifa['Weight']>=150) & (fifa['Weight']<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight)<200].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [.4, .1, 0, 0, .4]

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode)
plt.show()
```

### Box and Wisker Chart
Compare teams

```python
### The quickest way to figure out about the styling is usually a Google search instead of reading through documentations.


plt.style.use('default')     # Change style

plt.figure(figsize=(5, 8))   # Change figure size

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)
    
    # Change fill color
    box.set(facecolor='#e0e0e0') # For some reason need to set the patch_artist variable to True


plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```

```python
### Convert Height to an integer in inches
import re
import numpy as np

height_re = rf'(\d+)(\'+)(\d+)(\"*)'
height_feet_re = rf'(^\d+)'
height_inch_re = rf'(\d+$)'

def height_to_int(height):
    if isinstance(height, str):
        feet_value_string = re.search(height_feet_re, height).group()
        inch_value_string = re.search(height_inch_re, height).group()

        feet_value_int = int(feet_value_string)
        inch_value_int = int(inch_value_string)

        final_height_value_inch = feet_value_int * 12 + inch_value_int
#         print(final_height_value_inch)
        return(final_height_value_inch)
    else:
        return 0

fifa['Height (inch)'] = 0

### Create a new column of Height (inch)
for index, row in fifa.iterrows():
    fifa.loc[index, ['Height (inch)']] = height_to_int(row['Height'])
    

```

```python
height = fifa['Height (inch)']
weight = fifa['Weight']

plt.title('2018 FIFA Player Height-Weight dotplot')
plt.xlabel('Height (inch)')
plt.ylabel('Weight (lbs)')

plt.scatter(height, weight, marker='.')
plt.scatter(x=71, y=171, c='r', s=40)

plt.show()
```

### Finding the Weight Distribution of players Height=71

```python
import numpy as np

fifa.groupby(['Height (inch)']).mean()['Weight']
players_71inch_tall = fifa.loc[fifa['Height (inch)']==71]['Weight']


plt.title('Weight distribution of FIFA 2018 Players of 71" Tall')
plt.xlabel('Weight')
plt.ylabel('Count')

bins = np.arange(132, 203, 1).tolist()

plt.hist(players_71inch_tall, bins=bins)
plt.show()
```
