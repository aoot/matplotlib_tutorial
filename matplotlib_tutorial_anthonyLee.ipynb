```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]

### Resize the graph
plt.figure(figsize=(5, 3), dpi=300)

# plt.plot(x, y, label='2x', color='#ababab', linewidth=2, linestyle='--', marker='.', markersize=10, markeredgecolor='blue')

### Use shorthand notation
### fmt = '[color][marker][line]'
plt.plot(x, y, 'b^--', label='2x')


### Line number two
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.title('This is a graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

# Add a legend, requires label argument in the plotting
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution.)
plt.savefig('mygraph.png', dpi=1000)

# Show plot
plt.show()
```

### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0)) # pop() takes an index, 0 = popping off the first item.

# bars[0].set_hatch(patterns[0])
# bars[1].set_hatch(patterns[1])
# bars[2].set_hatch(patterns[2])


plt.figure(figsize=(6, 4))

plt.show()
```
